```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adsmission.csv')
```

```python
x = df['GRE Score']
y = df['TOEFL Score']

plt.scatter(x,y)
plt.show()
```

```python


x = df['GRE Score']
y = df['TOEFL Score']

plt.bar(x, y)
plt.xlabel('GRE Score')
plt.ylabel('TOEFL Score')
plt.title('GRE Score vs TOEFL Score')
plt.show()
```

```python
x = df['GRE Score']
y = df['Chance of Admit']

plt.bar(x, y)
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.title('GRE Score vs Chance of Admit')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['GRE Score'], df['Chance of Admit'], color='blue')
plt.title('GRE Score vs Chance of Admit')
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.show()
```

```python
avg_chance_by_rating = df.groupby('University Rating')['Chance of Admit'].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(avg_chance_by_rating['University Rating'], avg_chance_by_rating['Chance of Admit'], color='skyblue')
plt.title('Average Chance of Admit by University Rating')
plt.xlabel('University Rating')
plt.ylabel('Average Chance of Admit')
plt.show()
```
