```python
data2 = data.drop(columns=['Serial No.'])
data2.hist(bins=10, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
for i, column in enumerate(data2.columns, 1):
    plt.subplot(3, 3, i)
    sns.kdeplot(data2[column], color='pink')
    plt.title(f'Distribution of {column}')
```

```python
plt.figure(figsize=(15, 15))
for i, column in enumerate(data2.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data2[column], bins=10, color='yellow', kde=True, edgecolor='blue')
    plt.title(f'Distribution of {column}')
```
